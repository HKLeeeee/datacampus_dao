# 열 합치기

```python
merge3 = merge2[['번호', '교통량', 'SAMPLE_1','distance']]

# 열이름바꿔주기
merge3.columns = ['번호', '교통량', '환경영향평가점수', '건물까지거리(m)']
merge3
```
