```python
#명목형(Norminal)

def norminal_feature(df):
    nom_feature = [col for col in df.columns if df[col].dtypes == "object"]
    
    for col in nom_feature:
        df[col].value_counts().plot(kind='bar')
        plt.title(col)
        plt.show()
        
    return nom_feature
```

```python
#순서형(Ordinal)

def ordinal_feature(df):
    ord_feature = df.iloc[:,[4,5,6,12]]
    
    for col in ord_feature:
        df[col].value_counts().plot(kind='bar')
        plt.title(col)
        plt.show()
    
    return ordinal_feature

ordinal_feature()
```

```python
#수치형

def numerical_feature(df):
    df_info = df.iloc[:, 0:4]
    num_feature = list(set(df.columns) - set(df_info) - set(norminal_feature(df))\
                        - set(ordinal_feature(df)))
    num_feature = np.sort(num_feature)

    for col in num_feature:
        sns.distplot(df.loc[df[col].notnull(), col])
        plt.title(col)
        plt.show()
```
