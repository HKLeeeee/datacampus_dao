```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# df 생성 : 경사도 shp
file='/content/drive/MyDrive/Z_SIS_ASIT_SOILSLOPE_AREA.zip/Z_SIS_ASIT_SOILSLOPE_AREA.shp'
df = gpd.read_file(file, encoding='cp949')
df.info()
```

```python
# 경사도 labeling
df=df.replace(['기타','0-2%','2-7%','7-15%','15-30%','30-60%','60-100%'],['0','1','2','3','4','5','6'])

# dtype 숫자형으로 변경
df['SOILSLOPE']=pd.to_numeric(df['SOILSLOPE'], errors = 'raise')
df.info()
```

```python
# GeoDataFrame → shp
df.crs= "+proj=longlat +ellps=WGS84 +datum=WGS84 +no_defs"
df.to_file('/content/drive/MyDrive/Z_SIS_ASIT_SOILSLOPE_AREA_INT/Z_SIS_ASIT_SOILSLOPE_AREA_INT.shp', driver='ESRI Shapefile', encoding='euc-kr')
```
