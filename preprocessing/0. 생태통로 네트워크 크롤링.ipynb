```python
# 다운받은 크롬 드라이버 경로 수정
path = "../data/chromedriver/chromedriver.exe"
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
from selenium.webdriver.chrome.options import Options
```

```python
# 데이터별 리스트
date_li = []
facility_li = []
human_li = []
method_li = []
hiker_li = []
accuracy_li = []
efficiency_li = []
animal_li = []
improvements_li = []
bridge_name_li = []
bridge_type_li = []
road_li = []
facility2_li = []
install_li = []
year_li = []
geo_li_1_li = []
geo_li_2_li = []
address_li = []
```

```python
def save_csv():
    df = pd.DataFrame({
        '생태통로명':bridge_name_li,
        '주소':address_li,
        '도로':road_li,
        '관리기관':facility2_li,
        '설치기관':install_li,
        '생태통로타입':bridge_type_li,
        '준공년도':year_li,
        '생태통로좌표(위도)':geo_li_1_li,
        '생태통로좌표(경도)':geo_li_2_li,
        '모니터링기간':date_li,
        '수행기관':facility_li,
        '수행자':human_li,
        '모니터링방법':method_li,
        '등산객이용빈도':hiker_li,
        '모니터링신뢰도':accuracy_li,
        '생태통로효율성':efficiency_li,
        '이용확인종,이용빈도높은종,멸종위기종이용빈도':animal_li,
        '개선필요사항':improvements_li
    })
    df.to_csv('../data/main/생태통로 크롤링.csv', index=False)
```

```python
# 생태통로 모니터링 데이터 수집
def info_monitoring():
    data_li = []
    data_dict = {
        "date":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[1]/th/label",
        "date_val":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[1]/td",
        "facility":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[2]/th[1]/label",
        "facility_val":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[2]/td[1]",
        "human":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[2]/th[2]/label",
        "human_val":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[2]/td[2]",
        "method":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[3]/th/label",
        "method_val":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[3]/td",
        "hiker":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[4]/th/label",
        "hiker_val":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[4]/td",
        "accuracy":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[5]/th/label",
        "accuracy_val":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[5]/td",
        "efficiency":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[6]/th/label",
        "efficiency_val":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[6]/td",
        "animal":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[6]/th/label",
        "animal_val":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[7]/td",
        "improvements":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[7]/th/label",
        "improvements_val":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[8]/td"
        }
    for key, value in data_dict.items():
        data = driver.find_element(By.XPATH, value).text
        if data:
            data_li.append(data)
        else:
            data_li.append(" ")
    
    date_li.append(data_li[1])
    facility_li.append(data_li[3])
    human_li.append(data_li[5])
    method_li.append(data_li[7])
    hiker_li.append(data_li[9])
    accuracy_li.append(data_li[11])
    efficiency_li.append(data_li[13])
    animal_li.append(data_li[15])
    improvements_li.append(data_li[17])
```

```python
# 생태통로 정보 데이터 수집
def info_facility():
    data_li = []
    data_dict = {
        "bridge_name":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[1]/th/label",
        "bridge_name_val":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[1]/td",
        "address":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[2]/th[1]/label",
        "address_val":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[2]/td[1]",
        "roal":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[2]/th[2]/label",
        "road_val":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[2]/td[2]",
        "facility_name":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[3]/th[1]/label",
        "facility_name_val":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[3]/td[1]",
        "install_name":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[3]/th[2]/label",
        "install_name_val":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[3]/td[2]",
        "bridge_type":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[4]/th[1]/label",
        "bridge_type_val":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[4]/td[1]",
        "year":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[4]/th[2]/label",
        "year_val":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[4]/td[2]",
        "geo":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[5]/th/label",
        "geo_val_1":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[5]/td",
        "geo_val_2":"/html/body/div/div[2]/div/form/div[2]/table/tbody/tr[6]/td"
        }
    for key, value in data_dict.items():
        data = driver.find_element(By.XPATH, value)
        if data:
            data = data.text
            data_li.append(data)
        else:
            data = data.text
            data_li.append(None)
    
    bridge_name_li.append(data_li[1])
    address_li.append(data_li[3])
    road_li.append(data_li[5])
    facility2_li.append(data_li[7])
    install_li.append(data_li[9])
    bridge_type_li.append(data_li[11])
    year_li.append(data_li[13])
    geo_li_1_li.append(data_li[15])
    geo_li_2_li.append(data_li[16])
    
    print(len(bridge_name_li))
```

```python
# 페이지 이동 함수
def page_mover(start = 3, end = 13, semi_end = 11, step = True):
    xpath = "/html/body/div/div[5]/div/form/div/div[3]/table/tbody/tr/td[1]/div[3]"
    # 3 ~ 12
    for idx in range(start, end):
        re_xpath = xpath + ("/a[{}]".format(idx))
        current_page = driver.find_element(By.XPATH, re_xpath)
        current_page.click()
        print("next =>", idx-2)
        main_process(semi_end)
    if step:
        current_page = driver.find_element(By.XPATH, "/html/body/div/div[5]/div/form/div/div[3]/table/tbody/tr/td[1]/div[3]/a[13]")
        current_page.click()
```

```python
# 셀레니움 
def main_process(end):
    for i in range(1, end):
        xpath = '/html/body/div/div[5]/div/form/div/div[3]/table/tbody/tr/td[1]/div[2]/table/tbody/tr[{}]'.format(i)
        target = driver.find_element(By.XPATH, xpath)
        target.click()
    
        for j in range(1, 4):
            xpath_child = xpath + '/td[{}]'.format(j)
            target = driver.find_element("xpath", xpath_child).text
            # print(target)
        
        target = driver.find_element(By.XPATH, xpath)
        target.click()
        child_target_1 = driver.find_element(By.XPATH, child_link_1)
        child_target_1.click()
        driver.switch_to.window(driver.window_handles[1])
        info_facility()
        driver.close()
    
        driver.switch_to.window(driver.window_handles[0])
        
        child_target_2 = driver.find_element(By.XPATH, child_link_2)
        child_target_2.click()
        driver.switch_to.window(driver.window_handles[1])
        info_monitoring()
        driver.close()
        
        driver.switch_to.window(driver.window_handles[0])
        
        # print()
```

```python
options = Options()
options.add_argument('headless') #headless모드 - 브라우저가 뜨지 않고 실행

driver = webdriver.Chrome(path, options = options)

base_url = "https://www.nie-ecobank.kr/wildlifecrossing/mapservice/MapserviceList.do"
driver.get(base_url)
time.sleep(3)

child_link_1 = "/html/body/div/div[5]/div/form/div/div[3]/table/tbody/tr/td[3]/div/div/div[1]/div[1]/div/div[1]/div[10]/div/div[1]/div/ul/li[5]/a[1]"
child_link_2 = "/html/body/div/div[5]/div/form/div/div[3]/table/tbody/tr/td[3]/div/div/div[1]/div[1]/div/div[1]/div[10]/div/div[1]/div/ul/li[5]/a[2]"

# 페이지 당 생태통로 table
for i in range(1, 11):
    xpath = '/html/body/div/div[5]/div/form/div/div[3]/table/tbody/tr/td[1]/div[2]/table/tbody/tr[{}]'.format(i)
    target = driver.find_element(By.XPATH, xpath)
    target.click()
        
    # 생태통로 table 내 데이터
    for j in range(1, 4):
        xpath_child = xpath + '/td[{}]'.format(j)
        target = driver.find_element("xpath", xpath_child).text
#         print(target)
    # 생태통로 정보
    child_target_1 = driver.find_element(By.XPATH, child_link_1)
    child_target_1.click()
    driver.switch_to.window(driver.window_handles[1])
    info_facility()
    driver.close()
    
    # 창 전환
    driver.switch_to.window(driver.window_handles[0])
    
    # 생태통로 모니터링
    child_target_2 = driver.find_element(By.XPATH, child_link_2)
    child_target_2.click()
    driver.switch_to.window(driver.window_handles[1])
    info_monitoring()
    driver.close()
        
    # 창 전환
    driver.switch_to.window(driver.window_handles[0])

print("step = 0")
page_mover(4) # 2~10

print("step = 10")
page_mover(3) # 12~20

print("step = 20")
page_mover(3) # 22~30

print("step = 30")
page_mover(3) # 32~40

print("step = 40")
page_mover(3) # 42~50

print("step = 50")
page_mover(3, 6, 11)
page_mover(6, 7, 7, False)

print("finish!!")
```
